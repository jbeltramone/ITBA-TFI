### Persistence Model / Naive Forecasting

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.naive import NaiveForecaster
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split

%matplotlib inline
```

```python
SALTO_train, SALTO_test = temporal_train_test_split(SALTO, test_size=48)
```

```python
forecaster = NaiveForecaster(strategy="last", sp=7)
```

```python
fh = ForecastingHorizon(SALTO_test.index, is_relative=False)
#fh = ForecastingHorizon(np.arange(1, len(SALTO_test.index)), is_relative=True)
```

```python
# step 4: fitting the forecaster
forecaster.fit(SALTO_train)
```

```python
# step 5: querying predictions
SALTO_pred = forecaster.predict(fh)
```

```python
SALTO_test.plot(figsize=(16,4), style='.-', legend=True, label='SALTO_test')
SALTO_pred.plot(style='.-', legend=True, label='SALTO_pred')
```

```python
from sktime.performance_metrics.forecasting import MeanAbsoluteError, MeanSquaredError, MeanAbsoluteScaledError
```

```python
mae = MeanAbsoluteError()
mse = MeanSquaredError()
rmse = MeanSquaredError(square_root=True)
mase = MeanAbsoluteScaledError()
```

```python
error_metrics=[{'mae': mae(SALTO_test, SALTO_pred)}, 
               {'mse': mse(SALTO_test, SALTO_pred)}, 
               {'rmse': rmse(y_true=SALTO_test, y_pred=SALTO_pred)}, 
               {'mase': mase(y_true=SALTO_test, y_pred=SALTO_pred, y_train=SALTO_train)}]
```

## Reducción

```python
# Número de observaciones del salto
observations = len(SALTO)
```

```python
# Número de features a generar (Reduction)
features = 7
```

```python
# Listas de features y de target
X = [] # Matriz de features
y = [] # Vector de targets
```

```python
# Hago la reducción
for i in range(0, observations-features):
    X.append(SALTO[i:i+features]) # 'features' valores
    y.append(SALTO[i+features])   # Un target 
```

```python
# Convierto las listas en arreglos de Numpay
X = np.array(X)
y = np.array(y)
```

```python
len(X)
```

```python
# Train / test split
fh = 48 # Pasos (días) a predecir

# Features splitting
X_train = X[:len(X)-fh]
X_test = X[len(X)-fh:]

# Target splitting
y_train = y[:len(X)-fh]
y_test = y[len(X)-fh:]
```

```python
# Importo un regresor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
# Lista con los regresores
regresores = [RandomForestRegressor(), XGBRegressor(), KNeighborsRegressor(), LinearRegression()]
```

```python
# Creo una lista para las métricas
error_metrics=[]

# Itero sobre la lista de regresores
for regresor in regresores:
    
    # Entreno el regresor con los conjuntos de train
    regresor.fit(X_train, y_train)
    
    # Predigo con los valores de X_test
    y_pred = regresor.predict(X_test)
    
    # Calculo metras
    error_metrics.append( {
                           'Regresor': type(regresor).__name__, 
                           'mae': mae(y_test, y_pred), 
                           'mse': mse(y_test, y_pred),
                           'rmse': rmse(y_true=y_test, y_pred=y_pred), 
                           'mase': mase(y_true=y_test, y_pred=y_pred, y_train=y_train)
                          }
                        )
    
# Convierto la lista de métricas en un dataframe
metrics_df = pd.DataFrame.from_dict(error_metrics)    
```

```python
metrics_df.sort_values('mae')[['Regresor', 'mse', 'mae', 'rmse']].plot(figsize=(15,5), grid=True, kind='bar', x='Regresor')
```

## Grid Search

```python
import itertools
```

```python
# Importo un regresor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor
```

```python
param_grid = {  
    'fh': [28, 112, 336],
    'regresores': [
                   LinearRegression(), 
                   #RandomForestRegressor(criterion='absolute_error', max_depth=4), 
                   #XGBRegressor(), 
                   #KNeighborsRegressor(),
                   #SVR(epsilon=0.2),
                   #SGDRegressor()
                  ],
    'features': [56]
    }
```

```python
# Generate all combinations of parameters
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
```

```python
# Creo una lista para las métricas
error_metrics=[]
```

```python
# Creo una lista para los gráficos
plots=[]
```

```python
# Número de observaciones del salto
observations = len(SALTO)
```

```python
# Imprimo el número de ciclos
ciclos = len(all_params)
print('\n\nCiclos de Grid Search:', ciclos)
ciclo = 1

# Itero sobre la lista de parámetros
for params in all_params:

    # Imrimo el número del ciclo actual
    print('Ciclo', ciclo, 'de', ciclos)

    # Número de features a generar (Reduction)
    features = params['features']
    
    # Listas de features y de target
    X = [] # Matriz de features
    y = [] # Vector de targets
    
    # Hago la reducción
    for i in range(0, observations-features):
        X.append(SALTO[i:i+features]) # 'features' valores
        y.append(SALTO[i+features])   # Un target 
        
    # Convierto las listas en arreglos de Numpay
    X = np.array(X)
    y = np.array(y)    
    
    # Train / test split
    fh = params['fh'] # Pasos (días) a predecir
 
    # Features splitting
    X_train = X[:len(X)-fh]
    X_test = X[len(X)-fh:]

    # Target splitting
    y_train = y[:len(X)-fh]
    y_test = y[len(X)-fh:]
    
        
     # Regresor a utilizar
    regresor = params['regresores']
    
    # Entreno el regresor con los conjuntos de train
    regresor.fit(X_train, y_train)
    
    # Predigo con los valores de X_test
    y_pred = regresor.predict(X_test)
    
    # Convierto las y_'s de Numpy a Pandas series
    SALTO_train = pd.Series(y_train)
    SALTO_train.index = SALTO[:len(y_train)].index
    SALTO_test = pd.Series(y_test)
    SALTO_test.index = SALTO[-len(y_test):].index
    SALTO_pred = pd.Series(y_pred)
    SALTO_pred.index = SALTO[-len(y_test):].index
    
    
    # Ploteo los resultados
    titulo = type(regresor).__name__  + ',  features: ' + str(features) + ',  fh:'+ str(fh)  
    plots.append(plot_series(SALTO_train[-fh:], SALTO_test, SALTO_pred, 
                             labels=["SALTO_train", "SALTO_test", "SALTO_pred"], x_label=titulo)) 
    
    # Scatter Plot y_test vs y_pred
    #plt.scatter(y_test, y_pred)
    #plt.title(titulo)
    #plt.xlabel("y_test")
    #plt.ylabel("y_pred")
    #plt.show()
    
    
    # Calculo metricas
    error_metrics.append( {
                           'regresor': type(regresor).__name__,
                           'features': features,
                           'fh': fh,
                           'mae': mae(y_test, y_pred), 
                           'mse': mse(y_test, y_pred),
                           'rmse': rmse(y_true=y_test, y_pred=y_pred), 
                           'mase': mase(y_true=y_test, y_pred=y_pred, y_train=y_train)
                          }
                        )
    # Incremento el contador de ciclos
    ciclo = ciclo+1

print('\n\nFin Grid Search')
    
# Convierto la lista de métricas en un dataframe
metrics_df = pd.DataFrame.from_dict(error_metrics).sort_values(['fh', 'mae', 'features'])    
```
